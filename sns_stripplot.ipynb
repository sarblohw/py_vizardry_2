# STRIP Plot
- jitter: randomly plot each observation to segregate data points on a single observation
- goal of swarm and strip plot is to plot each observation as a circular dot
- in a swarm plot dots are arranged in vertical columns which helps look at the distribution of the data
- in a strip plot, the dots a randomly arranged, the density of data is figured out by the intensity of the colors or dots in that place

```python
import pandas as pd
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna(how='any').query('cylinders.isin([4,6,8])')
```

```python
cars.cylinders.value_counts()
```

```python
sns.stripplot(data=cars, x='weight');
```

```python
sns.stripplot(data=cars, x='weight', y='origin');
```

```python
sns.stripplot(data=cars, y='weight', x='origin');
```

```python
sns.stripplot(data=cars, y='weight', x='origin', hue='cylinders');
```

```python
sns.stripplot(data=cars, y='weight', x='origin', hue='cylinders', dodge=True);
```

```python
sns.stripplot(data=cars, x='weight', y='origin', hue='cylinders', dodge=True);
```

```python
sns.stripplot(data=cars, x='weight', y='origin', hue='cylinders', dodge=True, jitter=False);
```

```python
sns.stripplot(data=cars, x='weight', y='origin', hue='cylinders', dodge=True, jitter=0.05);
```

```python
sns.stripplot(data=cars, x='weight', y='origin', hue='cylinders', dodge=True, jitter=0.30);
#jitter - if the width of the column is considered as 1, then the jitter is +- the length from the center the data will be spread in
```

```python
from matplotlib import pyplot as plt
```

```python
from ipywidgets import interactive

def jitter_widget(jitter=0.1):
    sns.stripplot(data=cars, x='weight', y='origin', jitter=jitter)

interactive(jitter_widget, jitter=(0, 0.65, 0.05))
```

```python
sns.stripplot(data=cars, x='weight', y='origin', hue='cylinders', dodge=True, jitter=0.30, alpha=0.4, size=7);
```

```python
sns.stripplot(data=cars, x='weight', y='origin', hue='cylinders', dodge=True, jitter=0.30, alpha=0.8, size=3);
```

```python
sns.stripplot(data=cars, x='weight', y='origin', hue='cylinders', dodge=True, jitter=0.30, alpha=0.8, size=5, linewidth=1, edgecolor='black');
```

```python
sns.stripplot(data=cars, x='weight', y='origin', hue='cylinders', dodge=True, jitter=0.30, alpha=0.8, size=5, linewidth=1, edgecolor='black', marker='x');
```
