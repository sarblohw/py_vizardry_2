# SCATTER Plot
- Relational Plot

```python
import pandas as pd
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.info()
```

```python
diamonds = diamonds.loc[diamonds.cut.isin(['Premium', 'Good']) & diamonds.color.isin(['D', 'F', 'J'])].sample(n=100, random_state=22)
```

```python
diamonds.head(n=3)
```

```python
diamonds.shape
```

```python
sns.scatterplot(data=diamonds, x='carat', y='price');
```

```python
sns.scatterplot(data=diamonds, x='carat', y='price', marker='x');
```

```python
sns.scatterplot(data=diamonds, x='x', y='y');
```

```python
#semantice like hue, style or size of the markers can be used to add dimensions
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price', hue='cut'
);
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price', hue='cut',
    palette='pastel'
);
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price', hue='depth',
    palette='pastel'
);
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price', hue='cut', style='cut',
    palette='pastel', s=100
);
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price', hue='cut', style='cut', size='cut',
    palette='pastel'
);
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price', hue='cut', style='cut', size='cut',
    palette='pastel'
);
#palette and size can be fed with a list of colors/hexcodes or a list of variables in case of size
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price', hue='cut', style='cut', size='depth',
    palette='pastel'
);
#can use continuous variables also with the semantics
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price', hue='cut', style='color', size='depth',
    palette='pastel'
);
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price',
    palette='pastel', s=100, alpha=0.6
);
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price',
    palette='pastel', s=100, alpha=0.6, marker='*'
);
```

```python
sns.scatterplot(
    data=diamonds,
    x='carat', y='price',
    palette='pastel', s=100, alpha=0.6, marker='*', edgecolor='black'
);
#matplotlib scatter plot options can be used since this inherits from that
```
