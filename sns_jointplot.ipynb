# JOINT Plot
- Shows relationship b/w 2 variables and their individual distributions
- 3 plots in one

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
geyser = sns.load_dataset('geyser')
```

```python
geyser.head(3)
```

```python
geyser.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration');
```

```python
#you can choose to display scatter(default), kde, reg, hist, hex, resid
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', kind='kde');
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', kind='reg');
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', kind='hist');
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', kind='hex');
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', kind='resid');
```

```python
g = sns.jointplot(data=geyser, x='waiting', y='duration')
```

```python
type(g)
```

```python
g = sns.jointplot(data=geyser, x='waiting', y='duration')
g.plot_joint(sns.kdeplot, color='gray', levels=5)
```

```python
g = sns.jointplot(data=geyser, x='waiting', y='duration')
g.plot_joint(sns.kdeplot, color='gray', levels=5)
g.plot_marginals(sns.rugplot)
```

```python
#adding categorical data
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', hue='kind');
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', hue='kind', kind='kde');
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', color='g');
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', hue='kind', palette='autumn');
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', height=6, space=0.2, ratio=5);
#height is the height of the figure
#space is the space b/w the marginal figure and the center plot
#ratio is the ratio of size of the central plot to the marginal plot
```

```python
#joint_kws, marginal_kws
#the keywords are passed as per the kind of plot that exists in that area, joint or marginal
```

```python
sns.jointplot(
    data=geyser, x='waiting', y='duration',
    joint_kws=dict(marker=5)
);
```

```python
sns.jointplot(
    data=geyser, x='waiting', y='duration',
    kind='reg',
    joint_kws=dict(ci=None, order=2)
);
```

```python
sns.jointplot(
    data=geyser, x='waiting', y='duration',
    kind='reg',
    joint_kws=dict(ci=None, order=2),
    marginal_kws=dict(color='xkcd:golden')
);
```

```python
sns.jointplot(
    data=geyser, x='waiting', y='duration',
    kind='kde',
    marginal_kws=dict(linewidth=4, shade=True)
);
```
