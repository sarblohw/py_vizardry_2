# REG Plot
- Regression Plot to visualize Linear Regression Model

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds').sample(n=200, random_state=44)
```

```python
diamonds.shape
```

```python
sns.set_style('dark')
sns.set_palette('pastel')
```

```python
sns.regplot(data=diamonds, x='carat', y='price');
```

```python
sns.regplot(data=diamonds, x='carat', y='price', fit_reg=False);
```

```python
sns.regplot(data=diamonds, x='carat', y='price', scatter=False);
```

```python
#Remember causation is not implied here. It is just a relationship.
```

```python
#The shaded area is the confidence interval calculated using the bootstrapping method.
```

```python
sns.regplot(data=diamonds, x='carat', y='price', ci=None);
```

```python
#creating a discrete variable
cut_map = {
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Premium': 4,
    'Ideal': 5
}

diamonds['cut_value'] = diamonds.cut.map(cut_map)
```

```python
diamonds.cut_value.value_counts()
```

```python
diamonds['cut_value'] = pd.Categorical(diamonds.cut_value, categories=[1,2,3,4,5])
```

```python
import numpy as np
sns.regplot(data=diamonds, x='carat', y='price', x_jitter=0, x_estimator=np.mean);
```

#### Polynomial Regression

```python
sns.regplot(data=diamonds, x='carat', y='price', order=2);
```

#### Robust Regression

```python
x_example=[0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
y_example=[0.1, 0.8, 2.2, 2.7, 3.8, 4.5, 6.2, 6.8, 7.9, 9.4, 30.4]
```

```python
sns.regplot(x=x_example, y=y_example, ci=None, fit_reg=False);
```

```python
sns.regplot(x=x_example, y=y_example, ci=None, fit_reg=True);
```

```python
sns.regplot(x=x_example, y=y_example, robust=True);
```

### There are other than the linear regression model that can be plotted using the sns regplot
- Order (Polynomial): Higher powers working on the independent variables (not the coefficients)
- Robust: Robust to outliers
- Logistic: Variable Binary
- Lowess: Locally Weighted Scatter Plot Smoothing - Local regression to produce a very smoothed out curve
- xy partial: Compounding variables

```python
sns.regplot(x=diamonds.carat, y=diamonds.price, marker='d');
```

```python
sns.regplot(x=diamonds.carat, y=diamonds.price, scatter_kws={'s':50, 'alpha':0.5, 'color':'lightgray'});
```

```python
sns.regplot(x=diamonds.carat, y=diamonds.price, ci=None, line_kws={'lw':4, 'color':'lightgray', 'linestyle':'-.'});
```
