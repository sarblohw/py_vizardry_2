# ECDF Plot
- Empirical Cumulative Distribution Function
- Steps are made of vertical and horizontal lines
- - Vertical - the frequency of that observation on the x-axis
- - Horizontal - the distance between 2 consecutive observations on the x-axis
- This can show cumulative distribution. It however becomes harder to look at central tendencies and modes.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.ecdfplot(data=tips, x='tip');
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex');
```

```python
tips.tip.min()
```

```python
tips.tip.max()
#notice the range of the plot is the range of the dataset
```

```python
tips.tip.value_counts()
#notice the vertical lines at $2, $3 representing the frequency of those data points
```

```python
from matplotlib import pyplot as plt
```

```python
sns.ecdfplot(data=tips, x='tip')
plt.axvline(4, c='black');
#shows what proportion of the data was covered till we got to the $4 tips in ascending order from the lowest tip... here, it is about 80% or 0.8
```

```python
sns.ecdfplot(data=tips, y='tip')
plt.axvline(0.5, c='black');
#switched axis to show prop on x-axis, here the vertical line is drawn at 50% proportion or the mean
#we can interpret this as 50% of the tips are  done by around the $3 observation mark whe n they are arranged in an order
```

```python
tips.tip.median()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='time');
#observation: $2 lunch tips are common while $3 dinner is common
```

```python
sns.ecdfplot(data=tips, x='tip', stat='count');
#observation: 200 tips already received by the time we reach the $4 observation
```

```python
sns.ecdfplot(data=tips, x='tip', stat='count', complementary=True);
#complimentary switches axes so here we can count from the top observation backwards
#all tips from $10 to $2 count to about 170 tips
```

```python
sns.ecdfplot(data=tips, x='tip', hue='day', stat='count');
```

```python
sns.ecdfplot(data=tips, x='tip', hue='day', stat='count', complementary=True);
```

```python
sns.ecdfplot(data=tips, x='tip', hue='day', stat='proportion', complementary=False);
```

```python
sns.ecdfplot(data=tips, x='tip', weights='tip');
#using weight to weigh proportions
#here we can check the amount of total tip till that observation point
#by the time we reach $4 we have made about 70% of our total tip money
```

```python
sns.ecdfplot(data=tips, x='tip', weights='tip')
plt.axhline(0.5, c='black');
#we made 50%  of tip money by  about $3.5
```

```python
sns.ecdfplot(data=tips, x='tip', hue='day', palette='summer');
```

```python
sns.ecdfplot(data=tips, x='tip', hue='day', palette='summer', lw=3);
#plt functions used to style lines can be used like lw
```
