# SWARM Plot
- a circular dot for each data point
- the dots do not overlap and are at equal distance in a line
- Categorical scatter plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.dropna(how='any', subset=cars.columns, inplace=True)
```

```python
cars.head(3)
```

```python
sns.swarmplot(cars.mpg);
```

```python
cars = cars.loc[cars.cylinders.isin([4,6,8]), :]
```

```python
cars.sample(3)
```

```python
sns.swarmplot(data=cars, x='horsepower');
```

```python
sns.swarmplot(data=cars, x='origin', y='horsepower');
```

```python
sns.swarmplot(data=cars, x='origin', y='horsepower', hue='cylinders');
```

```python
sns.swarmplot(data=cars, x='origin', y='horsepower', hue='cylinders', dodge=True);
```

```python
usa = cars.query('origin == "usa"')
```

```python
usa.head(3)
```

```python
sns.boxplot(data=usa, x='cylinders', y='horsepower')
sns.swarmplot(data=usa, x='cylinders', y='horsepower');
```

```python
sns.boxplot(data=usa, x='cylinders', y='horsepower')
sns.swarmplot(data=usa, x='cylinders', y='horsepower', color='black');
```

```python
sns.boxplot(data=usa, x='cylinders', y='horsepower')
sns.swarmplot(data=usa, x='cylinders', y='horsepower', color='black', alpha=0.5);
```

```python
import numpy as np
```

```python
sns.boxplot(data=usa, x='cylinders', y='horsepower', whis=np.inf)
sns.swarmplot(data=usa, x='cylinders', y='horsepower', color='black', alpha=0.5);
```

```python
sns.violinplot(data=usa, x='cylinders', y='horsepower')
sns.swarmplot(data=usa, x='cylinders', y='horsepower', color='black', alpha=0.5);
```

```python
sns.violinplot(data=usa, x='cylinders', y='horsepower', scale='width')
sns.swarmplot(data=usa, x='cylinders', y='horsepower', color='black', alpha=0.5);
```

```python
sns.violinplot(data=usa, x='cylinders', y='horsepower', scale='width')
sns.swarmplot(data=usa, x='cylinders', y='horsepower', color='black', alpha=0.5, size=4);
```

```python
sns.violinplot(data=usa, x='cylinders', y='horsepower', scale='width', inner=None)
sns.swarmplot(data=usa, x='cylinders', y='horsepower', color='black', alpha=0.5, size=4);
```

```python
sns.violinplot(data=usa, x='cylinders', y='horsepower', scale='width', inner=None, color='xkcd:forest green')
sns.swarmplot(data=usa, x='cylinders', y='horsepower', color='black', alpha=0.5, size=3);
```
