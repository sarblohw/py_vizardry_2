# FACETGRID
It serves as the background for the catplot, relplot and displot.
- Small Multiple
- Pick out a categorical feature and plot for each category

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.sample(n=3)
```

```python
sns.set_style('white')
```

```python
sns.FacetGrid(data=penguins);
```

```python
sns.FacetGrid(data=penguins, col='island');
```

```python
g = sns.FacetGrid(data=penguins, col='island')
```

```python
type(g)
```

```python
sns.FacetGrid(data=penguins, row='island');
```

```python
g = sns.FacetGrid(data=penguins, col='island')
g.map(sns.histplot, 'flipper_length_mm')
```

```python
g = sns.FacetGrid(data=penguins, col='island')
g.map_dataframe(sns.histplot, x='flipper_length_mm')
#map_dataframe allows us to explicitly call out axis when plotting
```

```python
g = sns.FacetGrid(data=penguins, col='island')
g.map_dataframe(sns.histplot, y='flipper_length_mm')
```

```python
g = sns.FacetGrid(data=penguins, col='island')
g.map_dataframe(sns.scatterplot, x='bill_length_mm' , y='bill_depth_mm')
```
