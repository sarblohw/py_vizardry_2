# COUNT Plot
- Histogram for categorical variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds').query("clarity.isin(['SI1', 'VS2'])")
```

```python
diamonds.head(3)
```

```python
diamonds.shape
```

```python
sns.set_theme(style='darkgrid', palette='pastel')
```

```python
sns.countplot(data=diamonds, x='color');
```

```python
diamonds.color.value_counts()
```

```python
sns.countplot(data=diamonds, y='color');
```

```python
#while in a bar plot we get a summary stat/estimate and CI for a category, a count plot counts the observations
#options available for semantics and customizations are similar to those of a barplot
```

```python
sns.countplot(data=diamonds, x='cut');
```

```python
diamonds.dtypes
```

```python
diamonds.color.cat.categories
```

```python
diamonds.cut.cat.categories
```

```python
color_order = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
```

```python
sns.countplot(data=diamonds, x='color', order=color_order);
```

```python
diamonds.color.value_counts().sort_values(ascending=False)
```

```python
sns.countplot(data=diamonds, x='color', order=diamonds.color.value_counts().sort_values(ascending=False).index)
```

```python
sns.countplot(data=diamonds, x='color', order=diamonds.color.value_counts()[::-1].index);
```

```python
sns.countplot(data=diamonds, x='color', hue='clarity');
```

```python
sns.countplot(data=diamonds, x='color', hue='clarity', hue_order=['VS2', 'SI1']);
```

```python
sns.countplot(data=diamonds, x='color', color='skyblue');
```

```python
sns.countplot(data=diamonds, x='color', palette='twilight');
```

```python
sns.countplot(data=diamonds, x='color', palette='twilight', lw=4, ec='black', hatch='/');
#plt bar features will also work
```
