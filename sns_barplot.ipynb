# BAR Plot
- Categotical Estimate Plot
- One var is categorical and one an estimate/numerical, hence it is called a categotical estimate plot
- By default sns bar plot will calculate a mean of the categorical variable
- The confidence intervals are calculated at 95% by default using the Bootstrapping methodology

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head(n=3)
```

```python
sns.set_theme(style='darkgrid', palette='pastel')
```

```python
sns.barplot(data=penguins, x='species', y='body_mass_g');
```

```python
penguins.groupby('species').body_mass_g.agg('mean')
```

```python
sns.barplot(
    data=penguins,
    x='body_mass_g', y='species'
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g',
    ci=95
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g',
    ci=68
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g',
    ci=None
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g',
    ci=95,
    estimator=np.std
);
```

```python
penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g',
    ci=95,
    estimator=np.max
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g', hue='sex'
);
```

```python
sns.barplot(
    data=penguins,
    y='species', x='body_mass_g', hue='sex'
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g', hue='sex',
    order=['Gentoo', 'Chinstrap', 'Adelie'], hue_order=['Male', 'Female']
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g',
    color='c'
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g',
    color='c', errcolor='darkslateblue', errwidth=8
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g',
    color='c', errcolor='darkslateblue', errwidth=4, capsize=0.03
);
```

```python
sns.barplot(
    data=penguins,
    x='species', y='body_mass_g',
    palette='hot'
);
```
