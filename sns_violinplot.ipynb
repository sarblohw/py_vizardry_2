# VIOLIN Plot
- One part boxplot and one part and two parts kde plot
- Categorical Distribution Plot
- Inner part of the violin gives similar info to a box and whisker plot
- While a box plot shows outlier observations, the violin plot will show a kernel density estimation function of the spread

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head(3)
```

```python
cars = cars.loc[cars.cylinders.isin([4,6,8]), :]
```

```python
cars.head(3)
```

```python
cars.cylinders.value_counts()
```

```python
sns.set_style('dark')
sns.set_palette('pastel')
```

```python
sns.violinplot(data=cars, x='origin', y='mpg');
```

```python
sns.violinplot(cars.displacement);
```

```python
sns.violinplot(data=cars, x='cylinders', y='displacement');
```

```python
sns.violinplot(data=cars, x='cylinders', y='displacement', hue='origin');
```

```python
sns.violinplot(
    data = cars.query('origin == "japan" | origin == "europe"'),
    x = 'cylinders',
    y = 'mpg',
    hue = 'origin',
    split = True
)
plt.legend(loc=2);
#to utilize the split parameter, the hue argument should have exactly two values
```

```python
sns.violinplot(
    data = cars.query('origin == "japan" | origin == "europe"'),
    x = 'cylinders',
    y = 'mpg',
    hue = 'origin',
    split = True,
    inner = 'quartiles'
);
```

```python
sns.violinplot(
    data = cars.query('origin == "japan" | origin == "europe"'),
    x = 'cylinders',
    y = 'mpg',
    hue = 'origin',
    split = True,
    inner = None
);
```

```python
#by default the two split up pieces of the violin if integrated will have the same area, however we can scale it by count for example
sns.violinplot(
    data = cars.query('origin == "japan" | origin == "europe"'),
    x = 'cylinders',
    y = 'mpg',
    hue = 'origin',
    split = True,
    scale = 'count'
);
```

```python
sns.violinplot(
    data = cars.query('origin == "japan" | origin == "europe"'),
    x = 'cylinders',
    y = 'displacement',
    hue = 'origin',
    split = True,
    scale = 'count'
);
#here the width of the two halves of the violin is by number of observations due to the count
#note that here the plot is first splitting the data into 4 and 6 cylinders and then splitting by count
```

```python
sns.violinplot(
    data = cars.query('origin == "japan" | origin == "europe"'),
    x = 'cylinders',
    y = 'mpg',
    hue = 'origin',
    split = True,
    scale = 'count',
    scale_hue = False
);
#here the number of observations between the 4 cylinder and 6 cylinder cars can be compared
```

```python
sns.violinplot(
    x=cars.origin,
    y=cars.displacement,
    order=['japan', 'europe', 'usa']
);
```

```python
sns.violinplot(
    x=cars.origin,
    y=cars.displacement,
    order=['japan', 'europe', 'usa'],
    linewidth=3
);
```

```python
sns.violinplot(
    x=cars.origin,
    y=cars.displacement,
    order=['japan', 'europe', 'usa'],
    linewidth=3,
    bw=0.2
);
```

```python
sns.violinplot(
    x=cars.origin,
    y=cars.displacement,
    order=['japan', 'europe', 'usa'],
    linewidth=3,
    bw=0.2,
    cut=0.1
);
```
