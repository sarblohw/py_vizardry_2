# HEATMAP
- quantity of a cross tabular sort of data depicted by the depth of the color

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head(n=3)
```

```python
cars.groupby(['origin', 'cylinders'])['cylinders'].count()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
cars.groupby('origin').cylinders.value_counts().unstack(level=1)
```

```python
(cars.
 groupby('origin').
 cylinders.
 value_counts().
 unstack(level=1).
 fillna(value=0)
)
```

```python
origin_cyl = (cars.
 groupby('origin').
 cylinders.
 value_counts().
 unstack(level=1).
 fillna(value=0)
 )
```

```python
sns.heatmap(origin_cyl);
```

```python
sns.heatmap(origin_cyl.T, cmap='YlGnBu')
```

```python
cars.corr()
#by default this is the pearsons correlation
```

```python
sns.heatmap(cars.corr());
```

```python
sns.heatmap(origin_cyl, cmap='Blues');
```

```python
sns.heatmap(origin_cyl, cmap='RdBu');
```

```python
sns.heatmap(cars.corr(), cmap='RdBu', center=0);
```

```python
sns.heatmap(cars.corr(), cmap='RdBu', center=0, vmin=-1, vmax=+1);
```

```python
sns.heatmap(cars.corr(), cmap='YlGnBu', annot=True);
```

```python
sns.heatmap(origin_cyl, cmap='YlGnBu', annot=True, fmt='.0f');
```

```python
sns.heatmap(origin_cyl, cmap='YlGnBu', annot=True, fmt='.0f', annot_kws=dict(fontsize=16, fontweight='bold', fontfamily='serif', color='black'));
```

```python
sns.heatmap(cars.corr(), cmap='RdBu', linewidth=1);
```

```python
sns.heatmap(cars.corr(), cmap='RdBu', linewidth=1, linecolor='white');
```

```python
sns.heatmap(cars.corr(), cmap='RdBu', linewidth=1, linecolor='white', xticklabels=False, yticklabels=False);
#xticklabels and yticklabels can take a list of values we want as ticks or a boolean to turn em off/on
```

```python
sns.heatmap(cars.corr(), cmap='RdBu', square=True);
```
