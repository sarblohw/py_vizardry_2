# KDE Plot
- Kernel Density Estimation
- Non-Parametric way which means we will not assume any underlying distribution(Gaussian) so we don't have a mean/std to fit
- A kernel is added to each data point, then it is simply added together to get the kernel density estimation plot
- The kernel chosen at each data point can be any normal, cosine, triangular, etc. As long as there is enough data, it wouldn't matter

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head(3)
```

### Uni-variate KDE Plot

```python
sns.kdeplot(cars.horsepower);
```

```python
sns.kdeplot(cars.horsepower, shade=True);
```

```python
sns.kdeplot(data=cars, x='horsepower', hue='cylinders', shade=True)
```

```python
sns.kdeplot(data = cars, x = 'horsepower', bw_method='scott', shade = True);
#bw - bandwidth - lower bw will show mode jagged peaks whereas higher will smoothen out the curve, higher may even cause loss of modes
#scott/pearson/etc are methods of calculating bw
```

```python
#cumulative density
sns.kdeplot(data = cars, x = 'horsepower', hue = 'origin', cumulative = True);
```

```python
sns.kdeplot(cars.horsepower, cumulative = True);
```

### Bi-variate KDE Plot

```python
sns.kdeplot(data=cars, x='horsepower', y='weight');
```

```python
sns.kdeplot(data=cars, x='horsepower', y='mpg', n_levels=20);
```

```python
sns.kdeplot(data=cars, x='horsepower', y='mpg', n_levels=20, shade=True);
```

```python
sns.kdeplot(data=cars, x='horsepower', y='mpg', n_levels=20, shade=True, shade_lowest=False);
```

```python
sns.kdeplot(data=cars, x='horsepower', y='mpg', n_levels=20, shade=True, thresh=0.2, cbar=True);
```

```python
sns.kdeplot(data=cars.query('cylinders == 4 | cylinders == 6 | cylinders == 8'),
            x='horsepower', y='mpg', hue = 'cylinders',
            n_levels=20,
            shade=True, thresh=0.2, cbar=True);
```

```python
sns.kdeplot(data=cars.query('cylinders == 4 | cylinders == 6 | cylinders == 8'),
            x='horsepower', y='mpg', hue = 'cylinders',
            n_levels=20,
            shade=True, thresh=0.2);
```
