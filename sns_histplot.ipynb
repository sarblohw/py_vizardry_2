# HIST Plot

```python
import pandas as pd
import seaborn as sns
```

```python
sns.set_style('darkgrid')
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head(3)
```

```python
penguins.info()
```

```python
penguins.dropna(how='any', subset=penguins.columns, inplace=True)
```

```python
penguins.info()
```

```python
sns.histplot(data=penguins, x='bill_length_mm');
```

```python
sns.histplot(data=penguins, y='bill_length_mm');
```

```python
sns.histplot(data=penguins, x='bill_length_mm', kde=True);
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=20, kde=True);
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=[30,40,50,60], kde=True);
```

```python
sns.histplot(data=penguins, x='bill_length_mm', binwidth=10, binrange=(30,60));
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=20, kde=True, stat='count');
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=20, kde=True, stat='density');
#density - sum of the area of the bars will amount to 1
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=20, kde=True, stat='probability');
#probability - the height of all the bars should sum up to 1
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=10, kde=True, stat='probability', cumulative=True);
#probability and cumulative put together can give a sense of how the probability build up to 1
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=20, stat='probability', cumulative=True, element='step', fill=False);
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=20, stat='probability', cumulative=True, element='step', fill=True);
```

```python
sns.histplot(data=penguins, x='bill_length_mm', hue='species');
```

```python
sns.histplot(data=penguins, x='bill_length_mm', hue='species', multiple='stack');
```

```python
sns.histplot(data=penguins, x='bill_length_mm', hue='species', multiple='fill');
```

```python
sns.histplot(data=penguins, x='bill_length_mm', hue='species', multiple='dodge');
```

```python
sns.histplot(data=penguins, x='bill_length_mm', hue='species', multiple='layer', kde=True);
```

```python
sns.histplot(data=penguins, x='bill_length_mm', hue='species', element='step');
```

```python
sns.histplot(data=penguins, x='bill_length_mm', hue='species', element='poly');
```

### Bi Variate Histogram

```python
sns.histplot(data=penguins, x='bill_length_mm', y='bill_depth_mm');
```

```python
sns.histplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', cbar=True);
```

```python
sns.histplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species');
```

```python
sns.histplot(data=penguins, x='bill_length_mm', y='species', hue='species');
```

```python
sns.histplot(data=penguins, x='species', hue='sex', multiple='dodge');
```

```python
sns.histplot(data=penguins, x='species', hue='sex', multiple='dodge', shrink=0.8);
```

```python
sns.histplot(data=penguins, x='species', hue='sex', multiple='dodge', shrink=0.8, stat='density');
```

```python
sns.histplot(data=penguins, x='species', hue='sex', multiple='dodge', shrink=0.8, stat='probability');
```

```python
sns.histplot(data=penguins, x='species', color='red');
```

```python
sns.histplot(data=penguins, x='species', hue='sex', palette='bone', multiple='dodge', shrink=0.8);
```

```python
sns.histplot(x=penguins.bill_depth_mm, bins = 15, stat='count', kde=True, fill=False);
```
